```python
# Recarregar e examinar o dataset detalhadamente
print("=== ANÁLISE INICIAL DO DATASET ===")
print(f"Dimensões: {df.shape[0]} jogadores × {df.shape[1]} variáveis")
print("\n=== ESTRUTURA DOS DADOS ===")

# Entender o significado das colunas
print("\n📋 Principais variáveis identificadas:")
print("• Rk: Ranking")
print("• Player: Nome do jogador")
print("• Nation: Nacionalidade")
print("• Pos: Posição")
print("• Squad: Time")
print("• Age: Idade")
print("• MP: Matches Played (Jogos disputados)")
print("• Starts: Jogos como titular")
print("• Min: Minutos jogados")
print("• 90s: Número de jogos de 90 minutos")
print("• Gls: Gols")
print("• Ast: Assistências") 
print("• xG: Expected Goals (Gols esperados)")
print("• xAG: Expected Assisted Goals")
print("• CrdY/CrdR: Cartões amarelos/vermelhos")

# Verificar valores ausentes
print(f"\n📊 Valores ausentes:")
missing = df.isnull().sum()
print(missing[missing > 0])

# Verificar tipos de dados
print(f"\n🔍 Distribuição por tipos de dados:")
print(df.dtypes.value_counts())
```

```python
# === PREPARAÇÃO E LIMPEZA DOS DADOS ===

print("🧹 LIMPEZA DOS DADOS")

# 1. Tratar valores ausentes
print(f"Valores ausentes antes: {df.isnull().sum().sum()}")

# Para nacionalidade, idade e ano de nascimento, vamos remover as linhas (são apenas 4)
df_clean = df.dropna(subset=['Nation', 'Age', 'Born'])
print(f"Linhas removidas: {len(df) - len(df_clean)}")
print(f"Valores ausentes após limpeza: {df_clean.isnull().sum().sum()}")

# 2. Identificar e remover colunas duplicadas (com sufixo .1)
# Aparentemente são estatísticas por 90 minutos vs totais
print(f"\n📊 ENGENHARIA DE VARIÁVEIS")

# Vamos manter as principais colunas para análise
main_cols = ['Player', 'Nation', 'Pos', 'Squad', 'Age', 'MP', 'Starts', 'Min', '90s',
            'Gls', 'Ast', 'G+A', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR', 
            'xG', 'npxG', 'xAG', 'npxG+xAG', 'PrgC', 'PrgP', 'PrgR']

df_main = df_clean[main_cols].copy()

# 3. Criar novas variáveis para análise
print("Criando novas variáveis...")

# Eficiência de gols (gols por jogo)
df_main['Gols_por_jogo'] = df_main['Gls'] / df_main['MP'].replace(0, 1)

# Eficiência de assistências
df_main['Assist_por_jogo'] = df_main['Ast'] / df_main['MP'].replace(0, 1)

# Participação em gols (gols + assistências)
df_main['Participacao_gols'] = df_main['G+A'] / df_main['MP'].replace(0, 1)

# Eficiência de titularidade
df_main['Taxa_titularidade'] = df_main['Starts'] / df_main['MP'].replace(0, 1)

# Minutos por jogo
df_main['Min_por_jogo'] = df_main['Min'] / df_main['MP'].replace(0, 1)

# Disciplina (cartões por jogo)
df_main['Cartoes_por_jogo'] = (df_main['CrdY'] + df_main['CrdR']) / df_main['MP'].replace(0, 1)

# Performance vs expectativa (diferença entre gols reais e esperados)
df_main['Dif_gols_xG'] = df_main['Gls'] - df_main['xG']

print(f"✅ Dataset limpo: {df_main.shape[0]} jogadores × {df_main.shape[1]} variáveis")
print(f"Novas variáveis criadas: {df_main.shape[1] - len(main_cols)}")

# Verificar as primeiras linhas
df_main.head()
```

```python
# === ANÁLISE DESCRITIVA DETALHADA ===

print("📈 ANÁLISE DESCRITIVA")

# Estatísticas das variáveis numéricas principais
numeric_vars = ['Age', 'MP', 'Gls', 'Ast', 'xG', 'Min_por_jogo', 'Gols_por_jogo', 'Participacao_gols']
print("\n📊 Estatísticas descritivas das variáveis principais:")
print(df_main[numeric_vars].describe().round(3))

# Análise por posição
print(f"\n⚽ ANÁLISE POR POSIÇÃO")
print("Distribuição de jogadores por posição:")
pos_counts = df_main['Pos'].value_counts()
print(pos_counts)

print(f"\nDesempenho médio por posição (Gols por jogo):")
pos_performance = df_main.groupby('Pos')['Gols_por_jogo'].agg(['mean', 'std', 'count']).round(3)
print(pos_performance)

# Top performers
print(f"\n🏆 TOP PERFORMERS")
print("Top 10 artilheiros:")
top_scorers = df_main.nlargest(10, 'Gls')[['Player', 'Squad', 'Pos', 'Gls', 'xG', 'Gols_por_jogo']]
print(top_scorers.to_string(index=False))

print(f"\nTop 10 assistentes:")
top_assists = df_main.nlargest(10, 'Ast')[['Player', 'Squad', 'Pos', 'Ast', 'xAG', 'Assist_por_jogo']]
print(top_assists.to_string(index=False))
```

# 🎯 2. MODELAGEM ESTATÍSTICA - REGRESSÃO LINEAR

## Objetivos da Modelagem:
1. **Predizer gols** baseado em expected goals (xG) e outras variáveis
2. **Predizer performance geral** (participação em gols) baseado em características do jogador
3. **Avaliar fatores** que influenciam o desempenho dos jogadores

## Hipóteses a serem testadas:
- H1: Expected goals (xG) é um bom preditor de gols reais
- H2: Minutos jogados por partida influencia a eficiência de gols
- H3: A idade do jogador afeta sua performance
- H4: A posição do jogador é determinante para seu desempenho

```python
# === MODELO 1: PREDIÇÃO DE GOLS ===

print("🎯 MODELO 1: PREDIÇÃO DE GOLS BASEADO EM xG")

# Preparar dados para o modelo
# Remover jogadores com poucos minutos para ter dados mais significativos
df_model = df_main[df_main['Min'] >= 300].copy()  # Pelo menos 300 minutos jogados
print(f"Jogadores no modelo: {len(df_model)} (mínimo 300 minutos)")

# Variáveis para predição de gols
X1 = df_model[['xG', 'Min_por_jogo', 'Age']].copy()
y1 = df_model['Gls'].copy()

# Dividir em treino e teste
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=42)

# Treinar modelo
modelo_gols = LinearRegression()
modelo_gols.fit(X1_train, y1_train)

# Fazer predições
y1_pred = modelo_gols.predict(X1_test)

# Avaliar modelo
r2_1 = r2_score(y1_test, y1_pred)
rmse_1 = np.sqrt(mean_squared_error(y1_test, y1_pred))
mae_1 = mean_absolute_error(y1_test, y1_pred)

print(f"📊 RESULTADOS DO MODELO 1:")
print(f"R² Score: {r2_1:.4f}")
print(f"RMSE: {rmse_1:.4f}")
print(f"MAE: {mae_1:.4f}")

# Coeficientes
print(f"\n📈 COEFICIENTES:")
coef_names = ['xG', 'Min_por_jogo', 'Age']
for name, coef in zip(coef_names, modelo_gols.coef_):
    print(f"{name}: {coef:.4f}")
print(f"Intercepto: {modelo_gols.intercept_:.4f}")

# Análise detalhada com statsmodels para intervalos de confiança
X1_sm = sm.add_constant(X1)
modelo_sm1 = sm.OLS(y1, X1_sm).fit()
print(f"\n📋 RESUMO ESTATÍSTICO DETALHADO:")
print(modelo_sm1.summary())
```

```python
# === MODELO 2: PREDIÇÃO DE PERFORMANCE GERAL ===

print("\n" + "="*60)
print("🎯 MODELO 2: PREDIÇÃO DE PARTICIPAÇÃO EM GOLS")

# Variáveis para predição de performance geral (participação em gols)
X2 = df_model[['Age', 'Min_por_jogo', 'xG', 'xAG']].copy()
y2 = df_model['Participacao_gols'].copy()

# Dividir em treino e teste
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3, random_state=42)

# Treinar modelo
modelo_performance = LinearRegression()
modelo_performance.fit(X2_train, y2_train)

# Fazer predições
y2_pred = modelo_performance.predict(X2_test)

# Avaliar modelo
r2_2 = r2_score(y2_test, y2_pred)
rmse_2 = np.sqrt(mean_squared_error(y2_test, y2_pred))

print(f"📊 RESULTADOS DO MODELO 2:")
print(f"R² Score: {r2_2:.4f}")
print(f"RMSE: {rmse_2:.4f}")

# Coeficientes
print(f"\n📈 COEFICIENTES:")
coef_names2 = ['Age', 'Min_por_jogo', 'xG', 'xAG']
for name, coef in zip(coef_names2, modelo_performance.coef_):
    print(f"{name}: {coef:.4f}")

# === TESTES DE HIPÓTESES ===

print("\n" + "="*60)
print("🧪 TESTES DE HIPÓTESES E INTERVALOS DE CONFIANÇA")

# H1: Expected goals é um bom preditor de gols reais
correlation_xG_gols, p_value_h1 = pearsonr(df_model['xG'], df_model['Gls'])
print(f"\n📍 H1 - xG prediz gols reais:")
print(f"Correlação: {correlation_xG_gols:.4f}")
print(f"P-value: {p_value_h1:.2e}")
print(f"Resultado: {'✅ Aceita H1' if p_value_h1 < 0.05 else '❌ Rejeita H1'}")

# H2: Minutos por jogo influenciam eficiência de gols  
correlation_min_gols, p_value_h2 = pearsonr(df_model['Min_por_jogo'], df_model['Gols_por_jogo'])
print(f"\n📍 H2 - Minutos/jogo influenciam eficiência:")
print(f"Correlação: {correlation_min_gols:.4f}")
print(f"P-value: {p_value_h2:.4f}")
print(f"Resultado: {'✅ Aceita H2' if p_value_h2 < 0.05 else '❌ Rejeita H2'}")

# H3: Idade afeta performance
correlation_age_perf, p_value_h3 = pearsonr(df_model['Age'], df_model['Participacao_gols'])
print(f"\n📍 H3 - Idade afeta performance:")
print(f"Correlação: {correlation_age_perf:.4f}")
print(f"P-value: {p_value_h3:.4f}")
print(f"Resultado: {'✅ Aceita H3' if p_value_h3 < 0.05 else '❌ Rejeita H3'}")

# H4: Teste ANOVA para diferença de performance por posição
from scipy.stats import f_oneway

# Pegar principais posições
main_positions = df_model['Pos'].value_counts().head(5).index
groups = [df_model[df_model['Pos'] == pos]['Gols_por_jogo'] for pos in main_positions]
f_stat, p_value_h4 = f_oneway(*groups)

print(f"\n📍 H4 - Posição determina performance:")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value_h4:.4f}")
print(f"Resultado: {'✅ Aceita H4' if p_value_h4 < 0.05 else '❌ Rejeita H4'}")
```

```python
# === INTERVALOS DE CONFIANÇA E VALIDAÇÃO ===

print("\n" + "="*60)
print("📊 INTERVALOS DE CONFIANÇA (95%)")

# Modelo detalhado para intervalos de confiança
X1_sm = sm.add_constant(X1)
modelo_detalhado = sm.OLS(y1, X1_sm).fit()

# Intervalos de confiança para os coeficientes
conf_int = modelo_detalhado.conf_int()
coef_names_full = ['Intercepto', 'xG', 'Min_por_jogo', 'Age']

print(f"\n📈 INTERVALOS DE CONFIANÇA DOS COEFICIENTES:")
for i, name in enumerate(coef_names_full):
    lower, upper = conf_int.iloc[i, 0], conf_int.iloc[i, 1]
    coef_val = modelo_detalhado.params[i]
    print(f"{name}: {coef_val:.4f} [{lower:.4f}, {upper:.4f}]")

# Predições com intervalo de confiança
pred_summary = modelo_detalhado.get_prediction(X1_sm)
pred_df = pred_summary.summary_frame(alpha=0.05)

# === VISUALIZAÇÕES DOS MODELOS ===

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. Predições vs Valores Reais - Modelo 1
axes[0,0].scatter(y1_test, y1_pred, alpha=0.6, color='blue')
axes[0,0].plot([y1_test.min(), y1_test.max()], [y1_test.min(), y1_test.max()], 'r--', alpha=0.8)
axes[0,0].set_xlabel('Gols Reais')
axes[0,0].set_ylabel('Gols Preditos')
axes[0,0].set_title(f'Modelo 1: Predições vs Reais (R²={r2_1:.3f})')
axes[0,0].grid(True, alpha=0.3)

# 2. Resíduos - Modelo 1
residuos1 = y1_test - y1_pred
axes[0,1].scatter(y1_pred, residuos1, alpha=0.6, color='red')
axes[0,1].axhline(y=0, color='black', linestyle='--', alpha=0.8)
axes[0,1].set_xlabel('Valores Preditos')
axes[0,1].set_ylabel('Resíduos')
axes[0,1].set_title('Modelo 1: Análise de Resíduos')
axes[0,1].grid(True, alpha=0.3)

# 3. Predições vs Valores Reais - Modelo 2
axes[1,0].scatter(y2_test, y2_pred, alpha=0.6, color='green')
axes[1,0].plot([y2_test.min(), y2_test.max()], [y2_test.min(), y2_test.max()], 'r--', alpha=0.8)
axes[1,0].set_xlabel('Performance Real')
axes[1,0].set_ylabel('Performance Predita')
axes[1,0].set_title(f'Modelo 2: Predições vs Reais (R²={r2_2:.3f})')
axes[1,0].grid(True, alpha=0.3)

# 4. Performance por Posição (confirmando H4)
pos_data = [df_model[df_model['Pos'] == pos]['Gols_por_jogo'] for pos in main_positions]
axes[1,1].boxplot(pos_data, labels=main_positions)
axes[1,1].set_title('Performance por Posição (H4 confirmada)')
axes[1,1].set_ylabel('Gols por Jogo')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# === PREDIÇÕES FUTURAS ===
print(f"\n🔮 EXEMPLO DE PREDIÇÕES FUTURAS:")

# Exemplo: Jogador hipotético
exemplo_jogador = {
    'xG': 15.0,      # 15 expected goals na temporada
    'Min_por_jogo': 75.0,  # 75 minutos por jogo em média
    'Age': 25.0      # 25 anos
}

pred_gols = modelo_gols.predict([[exemplo_jogador['xG'], 
                                 exemplo_jogador['Min_por_jogo'], 
                                 exemplo_jogador['Age']]])[0]

print(f"Jogador com xG={exemplo_jogador['xG']}, {exemplo_jogador['Min_por_jogo']} min/jogo, {exemplo_jogador['Age']} anos:")
print(f"Predição de gols na temporada: {pred_gols:.1f}")
```

```python
# === RECARREGAR DADOS E CRIAR DASHBOARD SIMPLIFICADO ===
# (Necessário após reinstalação de bibliotecas)

# Reimportar bibliotecas essenciais
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from scipy.stats import pearsonr
import warnings
warnings.filterwarnings('ignore')

# Recarregar dados
df = pd.read_csv('dataset.csv')
df_clean = df.dropna(subset=['Nation', 'Age', 'Born'])

# Recriação rápida das variáveis principais
main_cols = ['Player', 'Nation', 'Pos', 'Squad', 'Age', 'MP', 'Starts', 'Min', '90s',
            'Gls', 'Ast', 'G+A', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR', 
            'xG', 'npxG', 'xAG', 'npxG+xAG', 'PrgC', 'PrgP', 'PrgR']

df_main = df_clean[main_cols].copy()
df_main['Gols_por_jogo'] = df_main['Gls'] / df_main['MP'].replace(0, 1)
df_main['Participacao_gols'] = df_main['G+A'] / df_main['MP'].replace(0, 1)
df_main['Min_por_jogo'] = df_main['Min'] / df_main['MP'].replace(0, 1)

# Filtrar jogadores ativos
df_model = df_main[df_main['Min'] >= 300].copy()

# Dashboard visual com matplotlib/seaborn
plt.style.use('default')
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('⚽ DASHBOARD DE ANÁLISE - JOGADORES PREMIER LEAGUE', fontsize=16, fontweight='bold')

# 1. Top 10 Artilheiros
top_scorers = df_model.nlargest(10, 'Gls')
axes[0,0].barh(range(len(top_scorers)), top_scorers['Gls'], color='skyblue')
axes[0,0].set_yticks(range(len(top_scorers)))
axes[0,0].set_yticklabels(top_scorers['Player'], fontsize=9)
axes[0,0].set_xlabel('Gols')
axes[0,0].set_title('🥅 Top 10 Artilheiros')
axes[0,0].grid(True, alpha=0.3)

# 2. Gols vs Expected Goals
axes[0,1].scatter(df_model['xG'], df_model['Gls'], alpha=0.6, color='blue')
axes[0,1].plot([0, df_model['xG'].max()], [0, df_model['xG'].max()], 'r--', alpha=0.8)
axes[0,1].set_xlabel('Expected Goals (xG)')
axes[0,1].set_ylabel('Gols Reais')
axes[0,1].set_title('🎯 Gols vs Expected Goals')
axes[0,1].grid(True, alpha=0.3)

# 3. Performance por Posição
main_positions = df_model['Pos'].value_counts().head(6).index
pos_data = [df_model[df_model['Pos'] == pos]['Gols_por_jogo'] for pos in main_positions]
bp = axes[0,2].boxplot(pos_data, labels=main_positions, patch_artist=True)
for patch in bp['boxes']:
    patch.set_facecolor('lightgreen')
axes[0,2].set_title('📊 Performance por Posição')
axes[0,2].set_ylabel('Gols por Jogo')
axes[0,2].tick_params(axis='x', rotation=45)

# 4. Distribuição de Idades
axes[1,0].hist(df_model['Age'], bins=15, color='orange', alpha=0.7, edgecolor='black')
axes[1,0].set_xlabel('Idade')
axes[1,0].set_ylabel('Frequência')
axes[1,0].set_title('📅 Distribuição de Idades')
axes[1,0].grid(True, alpha=0.3)

# 5. Modelo de Regressão - Resultado
X = df_model[['xG', 'Min_por_jogo', 'Age']]
y = df_model['Gls']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = LinearRegression().fit(X_train, y_train)
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)

axes[1,1].scatter(y_test, y_pred, alpha=0.6, color='purple')
axes[1,1].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', alpha=0.8)
axes[1,1].set_xlabel('Gols Reais')
axes[1,1].set_ylabel('Gols Preditos')
axes[1,1].set_title(f'🤖 Modelo de Regressão (R² = {r2:.3f})')
axes[1,1].grid(True, alpha=0.3)

# 6. Resumo de Estatísticas
axes[1,2].text(0.1, 0.9, '📈 ESTATÍSTICAS PRINCIPAIS', fontsize=12, fontweight='bold', transform=axes[1,2].transAxes)
axes[1,2].text(0.1, 0.8, f'Total de jogadores: {len(df_model)}', fontsize=10, transform=axes[1,2].transAxes)
axes[1,2].text(0.1, 0.7, f'Correlação xG-Gols: {df_model["xG"].corr(df_model["Gls"]):.3f}', fontsize=10, transform=axes[1,2].transAxes)
axes[1,2].text(0.1, 0.6, f'Média de gols: {df_model["Gls"].mean():.1f}', fontsize=10, transform=axes[1,2].transAxes)
axes[1,2].text(0.1, 0.5, f'Média de idade: {df_model["Age"].mean():.1f} anos', fontsize=10, transform=axes[1,2].transAxes)

# Top 3 times
top_teams = df_model.groupby('Squad')['Gls'].sum().nlargest(3)
axes[1,2].text(0.1, 0.4, '🏆 TOP 3 TIMES (total gols):', fontsize=10, fontweight='bold', transform=axes[1,2].transAxes)
for i, (team, goals) in enumerate(top_teams.items()):
    axes[1,2].text(0.1, 0.3-i*0.05, f'{i+1}. {team}: {int(goals)}', fontsize=9, transform=axes[1,2].transAxes)

axes[1,2].set_xlim(0, 1)
axes[1,2].set_ylim(0, 1)
axes[1,2].axis('off')

plt.tight_layout()
plt.show()
```
